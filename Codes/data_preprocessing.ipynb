```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DataLoader
G = pd.read_csv('../2022_Problem_C_DATA/LBMA-GOLD.csv')
B = pd.read_csv('../2022_Problem_C_DATA/BCHAIN-MKPRU.csv')
df = B.join(G.set_index('Date'), on='Date', how='inner')
df.head()
print(df)
```

```python
# 5 days for one time extract the characterstic
gold = {'date':[],'high':[],'low':[],'open':[],'close':[]}
bitcoin = {'date':[],'high':[],'low':[],'open':[],'close':[]}
time_step_length = 5
for start in range(0,df.shape[0],time_step_length):
    end = start+time_step_length-1
    gold['date'].append(df.iloc[start,0])
    if(np.isnan(df.iloc[start,2])):
        gold['open'].append(df.iloc[start-1,2])
    else:
        gold['open'].append(df.iloc[start,2])
         
    if(np.isnan(df.iloc[end,2])):
        gold['close'].append(df.iloc[end-1,2])
    else:
        gold['close'].append(df.iloc[end,2])
    gold['high'].append(df.iloc[start:end+1,2].max(axis=0))
    gold['low'].append(df.iloc[start:end+1,2].min(axis=0))
    

    bitcoin['date'].append(df.iloc[start,0])
    bitcoin['open'].append(df.iloc[start,1])
    if(np.isnan(df.iloc[start,1])):
        print("nan detect at start")
    #     bitcoin['open'].append(df.iloc[start-1,1])
    # else:
    #     bitcoin['open'].append(df.iloc[start,1])
         
    if(np.isnan(df.iloc[end,1])):
        print("nan detect at end")
    #     bitcoin['close'].append(df.iloc[end-1,1])
    # else:
    bitcoin['close'].append(df.iloc[end,1]) 
    bitcoin['high'].append(df.iloc[start:end+1,1].max(axis=0))
    bitcoin['low'].append(df.iloc[start:end+1,1].min(axis=0))

gold = pd.DataFrame(gold)
bitcoin  = pd.DataFrame(bitcoin)
gold.to_csv("./poloniex_data/gold.csv",index=0)
bitcoin.to_csv("./poloniex_data/bitcoin.csv",index=0)
```
